```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import stemgraphic
import scipy
from scipy import stats
```

```python
passmark = 50
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
#female stuents
df_female = df[df["gender"]=="female"]
#male students
df_male = df[df["gender"]=="male"]
```

```python
df['Total_Marks'] = df['math score']+df['reading score']+df['writing score']
```

```python
df_female['Total_MarksFemale'] = df['math score']+df['reading score']+df['writing score']
```

```python
df_male['Total_MarksMale'] = df['math score']+df['reading score']+df['writing score']
```

```python
df_female['Math_PassStatusFemale'] = np.where(df_female['math score']<passmark, 'fF', 'fP')
df_female.Math_PassStatusFemale.value_counts()
```

```python
df_male['Math_PassStatusMale'] = np.where(df_male['math score']<passmark, 'mF', 'mP')
df_male.Math_PassStatusMale.value_counts()
```

```python
df['Math_PassStatus'] = np.where(df['math score']<passmark, 'F', 'P')
df.Math_PassStatus.value_counts()
```

```python
df_female['Reading_PassStatusFemale'] = np.where(df_female['reading score']<passmark, 'fF', 'fP')
df_female.Reading_PassStatusFemale.value_counts()
```

```python
df_male['Reading_PassStatusMale'] = np.where(df_male['reading score']<passmark, 'mF', 'mP')
df_male.Reading_PassStatusMale.value_counts()
```

```python
df['Reading_PassStatus'] = np.where(df['reading score']<passmark, 'F', 'P')
df.Reading_PassStatus.value_counts()
```

```python
df_female['Writing_PassStatusFemale'] = np.where(df_female['writing score']<passmark, 'fF', 'fP')
df_female.Writing_PassStatusFemale.value_counts()
```

```python
df_male['Writing_PassStatusMale'] = np.where(df_male['writing score']<passmark, 'mF', 'mP')
df_male.Writing_PassStatusMale.value_counts()
```

```python
df['Writing_PassStatus'] = np.where(df['writing score']<passmark, 'F', 'P')
df.Writing_PassStatus.value_counts()
```

```python
df_female['OverAll_PassStatusFemale'] = df_female.apply(lambda x : 'fF' if x['Math_PassStatusFemale'] == 'fF' or x['Reading_PassStatusFemale'] == 'fF' or x['Writing_PassStatusFemale'] == 'fF' else 'fP', axis =1)
df_female.OverAll_PassStatusFemale.value_counts()
```

```python
df_male['OverAll_PassStatusMale'] = df_male.apply(lambda x : 'mF' if x['Math_PassStatusMale'] == 'mF' or x['Reading_PassStatusMale'] == 'mF' or x['Writing_PassStatusMale'] == 'mF' else 'mP', axis =1)
df_male.OverAll_PassStatusMale.value_counts()
```

```python
df['OverAll_PassStatus'] = df.apply(lambda x : 'F' if x['Math_PassStatus'] == 'F' or x['Reading_PassStatus'] == 'F' or x['Writing_PassStatus'] == 'F' else 'P', axis =1)
df.OverAll_PassStatus.value_counts()
```
